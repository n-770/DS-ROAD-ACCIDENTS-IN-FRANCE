```python
import pandas as pd
```

```python
# -------------------------------------------------------------------------------------------------
# 1. read csv files
# -------------------------------------------------------------------------------------------------
df_charac_2019 = pd.read_csv('../data/gov_charac_2019.csv', sep=';',encoding='latin1')
df_charac_2020 = pd.read_csv('../data/gov_charac_2020.csv', sep=';',encoding='latin1')
df_charac_2021 = pd.read_csv('../data/gov_charac_2021.csv', sep=';',encoding='latin1')
df_charac_2022 = pd.read_csv('../data/gov_charac_2022.csv', sep=';',encoding='latin1')
df_charac_2023 = pd.read_csv('../data/gov_charac_2023.csv', sep=';',encoding='latin1')
df_charac_2024 = pd.read_csv('../data/gov_charac_2024.csv', sep=';',encoding='latin1')
```

```python
print('2019:', df_charac_2019.shape)
print('2020:', df_charac_2020.shape)
print('2021:', df_charac_2021.shape)
print('2022:', df_charac_2022.shape)
print('2023:', df_charac_2023.shape)
print('2024:', df_charac_2024.shape)

display(df_charac_2019.head(3))
display(df_charac_2020.head(3))
display(df_charac_2021.head(3))
display(df_charac_2022.head(3))
display(df_charac_2023.head(3))
display(df_charac_2024.head(3))
```

```python
# -------------------------------------------------------------------------------------------------
# 2. rename differing column names
# -------------------------------------------------------------------------------------------------
df_charac_2022 = df_charac_2022.rename(columns={'Accident_Id': 'Num_Acc'})
```

```python
# -------------------------------------------------------------------------------------------------
# 3. check dtypes over the year
# -------------------------------------------------------------------------------------------------
# List of years and corresponding DataFrames
years = list(range(2019, 2024))
dfs = {year: globals()[f'df_charac_{year}'] for year in years}

# Collect column-dtype pairs per year
records = []
for year, df in dfs.items():
    for col in df.columns:
        records.append({'Year': year, 'Column': col, 'Dtype': df[col].dtype.name})

# Create DataFrame from records
df_summary = pd.DataFrame(records)

# Crosstab: Column vs. Year, showing dtype
crosstab = pd.crosstab(df_summary['Column'], df_summary['Year'], values=df_summary['Dtype'], aggfunc='first')

crosstab
```

```python
# -------------------------------------------------------------------------------------------------
# 4. Concat yearly files to one df
# -------------------------------------------------------------------------------------------------
df_charac_2019_2024 = pd.concat([df_charac_2019, df_charac_2020, df_charac_2021, df_charac_2022, df_charac_2023, df_charac_2024], axis=0)

print(df_charac_2019_2024.shape)
print(df_charac_2019_2024.info())
```

```python
# -------------------------------------------------------------------------------------------------
# 5. export final dataframe to joblib
# -------------------------------------------------------------------------------------------------
from joblib import dump

dump(df_charac_2019_2024, '1.0-leibold-data-exploration_accidents.joblib')
```
